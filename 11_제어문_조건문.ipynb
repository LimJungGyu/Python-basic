어떤 특정한 조건에서만 코드가 작동되게 하려면 어떻게 할까요? 파이썬의 경우에는 if, elif, else 를 이용하여 코드를 제어합니다. n가지 조건이 나열되어 있을 때 아래와 같이 if, elif, else를 배분합니다.

```
if 조건1:
  조건1 만족시 작동
elif 조건2:
  조건2 만족시 작동
...
elif 조건n-1:
  조건n-1 만족시 작동
else:
  조건 n 만족시 작동
```


### **(1) if문의 사용법**









#### **1. if**
간단한 if 문을 작성해 봅시다. if문 하나만으로도 코드를 작성할 수 있습니다. 

```
a = 2

if a>0:
  print("0보다 큽니다")
```


````python
a = 2

if a>0:
  print("0보다 큽니다")
  print("if문의 영향받는 출력")
print("if문의 영향을 받지않는 출력")
print("")
if a<0:
  print("0보다 큽니다")
  print("if문의 영향받는 출력")
print("if문의 영향을 받지않는 출력")
````

####**2. else**

만약 주어진 조건을 만족하지 않을 시에 특정 코드를 작동시키고 싶으면 조건을 추가할 수 있습니다. 두가지 조건이 된다면 첫번째 조건은 if문으로, 마지막 조건은 else문이 됩니다.

```
a = 2

if a>0:
  print("0보다 큽니다.")
else:
  print("0보다 같거나 작습니다.")
```







````python
a = 2

if a>0:
  print("0보다 큽니다.")
else:
  print("0보다 같거나 작습니다.")
````

####**3. elif**

만약 조건이 세개 이상으로 나뉘면 elif 문을 추가해줍니다. 이제 조건은 더 세분화 되었습니다.

```
a = 2

if a>0:
  print("0보다 큽니다.")
elif a==0:
  print("0입니다")
else:
  print("0보다 작습니다.")
```








````python
a = 2

if a>0:
  print("0보다 큽니다.")
elif a==0:
  print("0입니다")
else:
  print("0보다 작습니다.")
````

### **(2) 삼항 연산자**

조건문과 비슷한 기능을 하는 연산자입니다. 똑같이 조건문이지만 다른 format으로 쓸 수 있구나 라고 알고 가시면 됩니다.


```
# 조건문
if 조건: -(1), (2)
  조건이참일때코드 - (3)
else: -(4)
  조건거짓일때코드 -(5)

# 삼항 연산자
조건참일때코드 if 조건 else 조건거짓일때코드
    (3)   (1) (2) (4)     (5)
```

위 (1)-1 에서 작성한 코드를 삼항 연산자로 변환해 봅니다.

````python
a = 2

if a>0:
  print("0보다 큽니다")
else :
  print("0보다 작습니다")
print("0보다 큽니다") if a>0 else print ("dd")
````

### **(3) 여러가지 조건문**





####**1. 복잡한 조건**

여기서 말하는 복잡한 조건은 and나 or로 연결된 다중 조건을 말합니다. 대부분의 경우에 조건이 그렇게 복잡하지는 않습니다. 아래는 3의 배수 조건입니다.

```
a = 12

if a>=3 and a%3==0:
  print(a, "는 3의 배수입니다")
```

and 나 or 등 논리 연산자에 대한 자세한 설명은 chapter05-불료형에서 다시 확인할 수 있습니다. 

````python
a = 12

if a>=3 and a%3==0:
  print(a, "는 3의 배수입니다")
````

####**2. 이중 if 문**

if문 내에 if 문이 들어가 있을 때를 얘기합니다. if-elif-else 모두 이중구조로 사용할 수 있습니다. 필요시 문제에 맞게 적절히 if문을 짜도록 합니다.

```
a = 12

if a<100:
  if a%2==0:
    print("짝수입니다.")
else: 
  "입력숫자초과"
```

 

````python
a = 12

if a<100:
  if a%2==0:
    print("짝수입니다.")
else: 
  "입력숫자초과"
````

\####**예제 04**

0에서 100까지의 정수 n을 입력받습니다. 만약 잘못된 정수를 입력하면 에러 메시지를 보내고 종료됩니다. 제대로 입력하였으면 정수 n에 10을 빼고 출력합니다. 만약 10을 뺀 값이 30보다 작으면 30을 출력하고 70보다 크면 70을 출력하세요.


```
>> limit_nums(50)
40
>> limit_nums(35)
30
>> limit_nums(95)
70
>> limit_nums(200)
잘못된 숫자입니다.
```



````python
def limit_nums(n):
  if n<100:
    if n-10<30:
      print("30")
    elif n-10>70:
      print("70")
    else:
      print(n-10)
  else:
    print("에러 잘못된 정수입력")
limit_nums(50)
limit_nums(35)
limit_nums(95)
limit_nums(200)

````

* 음료수 종류를 입력받고 가격을 안내합니다.

* 금액이 충분하면 음료수와 거스름돈을 반환하는 메시지를 출력합니다.

````python
coffee = {"아메리카노":1000,"라떼":1200,"밀크티":1800}
print(coffee)
order = input("어떤 것으로 주문하시겠습니까?:")
print("{0}은 {1}원입니다".format(order,coffee[order]))
cash = int(input("금액 입력받습니다"))
if coffee[order]<=cash :
  print("주문하신 {0}가 나왔습니다 거스름돈은 {1}원입니다".format(order,cash-coffee[order]))
else:
  print("주문하신 {0}은 금액이 불충분합니다 {1}원 돌려드립니다".format(order,cash))
````

####**예제 09**

두 수를 비교하는 함수를 만듭니다.



```
>> comp_ab(1,2)
1<2
>> comp_ab(2,1)
2>1
>> comp_ab(1,1)
1==1
```



````python
def comp_ab(a,b):
  if a>b:
    print("{0}>{1}".format(a,b))
  elif a<b:
    print("{0}<{1}".format(a,b))
  else:
    print("{0}=={1}".format(a,b))
comp_ab(1,2)
comp_ab(2,1)
comp_ab(1,1)
````

####**예제 11**

####**예제 14**

두참가자에게 투표를 한 문자열이 주어집니다. 1번과 2번중 더 높은 투표를 받은 사람이 번호를 출력하세요. 만약 투표수가 동일하면 "draw"를 출력합니다.

```
>> who_win("122211")
draw
>> who_win("11121")
1
>> who_win("211222212")
2
```

````python
def who_win(vote):
  a=list(vote).count("1")
  b=list(vote).count("2")
  if a>b:
    print("A",a,b)
  elif a<b:
    print("B",a,b)
  else:
    print("Draw",a,b)
who_win("122211")
who_win("11121")
who_win("211222212")
````
