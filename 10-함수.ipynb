###**(3) 매개변수와 인수**




매개변수와 인수는 함수의 인풋이라고 볼 수 있습니다. 파이썬 함수에는 인풋과 아웃풋이 존재하지 않을 수도 있고, 하나 또는 그 이상도 가능합니다.

```
def plus(input1, input2):

```


#### **2. 인수 (argument)**

```

```

````python
plus(1, 2)
````

### **(4) 함수의 반환**

함수에 인풋과 마찬가지로 아웃풋도 존재합니다. 만약 함수를 통해 반환하는 값을 원한다면 **return** 을 이용합니다.

```
def plus(input1, input2):
  plus_value = input1+input2
  return plus_value
```





````python
def plus(input1, input2):
  plus_value = input1+input2
  return plus_value
````


이제 이 함수는 하나의 return 값을 가집니다. 따라서 이 함수를 사용할 때는 아래와 같이 적어줍니다.


```
a = plus(5, 5)
print(a)
```

````python
a=plus(5,5)
print(a,type(a))
````

여러개의 반환 값을 원한다면 아래와 같이 작성합니다.


```
def plus_minus(input1, input2):
  plus_value = input1+input2
  minus_value = input1-input2
  return plus_value, minus_value
```

````python
def plus_minus(input1, input2):
  plus_value = input1+input2
  minus_value = input1-input2
  return plus_value, minus_value
````

함수를 사용할 때는 아래와 같이 써줍니다.

```
a = plus_minus(5, 5)
print(a)
```

````python
a = plus_minus(5, 5)
print(a,type(a))
````


a 의 값이 튜플 형식으로 지정되어 있는 것을 볼 수 있습니다. 따라서 아래와 같이 함수를 사용하기도 합니다.


```
a_pl, a_mi = plus_minus(5,5)
print(a_pl, a_mi)
```



````python
a_pl, a_mi = plus_minus(5,5)
print(a_pl, a_mi)
````

### **(5) 여러가지 매개변수**





#### **1. 디폴트 값이 있는 매개변수** 

특정 인수의 디폴드 값을 아래와 같이 지정 할 수 있습니다. 이 때 디폴트 값이 있는 매개변수는 항상 마지막에 작성해야 하며 그렇지 않으면 에러 메시지가 출력됩니다.

```
def plus(input1, input2, bias=0):
  print(input1+input2+bias)
```

```
def plus(input1, bias=0, input2):
  print(input1+input2+bias)
```


````python
def plus(input1,input2,bias=0):   ##옵션은 맨마지막에 넣어야 오류가 안난다
  print(input1+input2+bias)

def plus(input1, bias=0, input2):  #이거 오류남
  print(input1+input2+bias)
````

디폴트로 지정되어 있는 값은 함수 실행시 따로 지정하지 않아도 함수가 작동되며 디폴트값을 조정 하고 싶다면 아래와 같이 함수를 작성합니다.

```
plus(input1, input2)
plus(input1, input2, bias=-1)
```




````python
plus(1,3)  #디폴트값이 지정되어 있기때문에 2개만 넣어도 3개째는 디폴트값으로 실행
plus(1,3, bias=-1)  #디폴트 변경
````

#### **2. args**

인풋의 개수가 지정되어 있지 않는 함수를 만들어야 할 때가 있습니다. 이럴 때는 args 를 사용합니다. 사용시에는 앞에 별 하나 * 를 붙입니다.

```
def many_input(*args):
  print(args)

many_input(1,2,3,4)
```




````python
def many_input(*args):
  print(args) #튜플 형태의 변수를 다룰 수 있다

many_input(1,2,3,4)
````

#### **3. kwargs**

키워드 파라미터 kwargs는 딕셔너리를 출력합니다. kwargs 를 쓸 때는 앞에 별 두개 ** 를 붙여서 사용합니다.

```
def my_func(**kwargs):
  print(kwargs)

my_func(a=1, b=2, c=3)
```




````python
def my_func(**kwargs):
  print(kwargs)  #딕셔너리 형태의 변수를 사용

my_func(a=1,b=2,c=3) #따옴표 넣지않아도 문자인식한다 오히려 따옴표를 넣으면 안된다 오류남
````

### **(6) 지역변수와 전역변수**

전역변수는 프로그램 전체에서 사용되는 변수이고 지역변수는 특정 코드 블록 내에서만 사용되는 변수입니다. 함수 내에서 코드를 작성하고 변수를 지정하면 지역변수가 생성됩니다. 아래 코드에서 a,b,c는 전역변수이고 함수내에 있는 a,b,c는 지역변수입니다. 함수내의 a,b,c,d는 전역변수에 영향을 미치지 않습니다. 

```
a = 1
b = 1
c = 1

def my_func(a, b):
  c = a+b
  d = a-b
  return c

my_out = my_func(5,5)
print(a,b,c)
```

````python
a = 1
b = 1
c = 1

def my_func(a, b):
  c = a+b
  d = a-b
  return c

my_out = my_func(5,5)
print(a,b,c)
````

함수 실행시에는 반환 값이 전역변수로 지정됩니다.

```
c = my_func(5,5)
print(a,b,c)
```

````python
c = my_func(5,5)
print(a,b,c)
````

함수내에서 선언했던 모든 변수들은 함수 밖에서 실행되지 않습니다.

```
print(d)
```

````python
print(d)
````

만약 함수내 지역변수를 전역변수로 지정하고 싶다면 **global** 을 사용합니다.

```
def my_func(a, b):
  global d
  c = a+b
  d = a-b
  return c

print(d)
```

````python
def my_func(a, b):
  global d
  c = a+b
  d = a-b
  return c
c= my_func(0,5)
print(d)  #global로 지역변수를 전역변수로 만들어서 오류가 나지 않는다 
````

####**예제 09**

무작위 숫자들을 입력하면 중복을 제거한 숫자들의 개수와 최소값, 최대값을 알려주는 함수 numbers() 를 작성하세요. 또한 최소값과 최대값을 반환하도록 하세요



```
>> mn, mx = numbers(1,3,2,1,1)
총 5가지 숫자가 들어왔습니다. 최소값 1, 최대값 3 입니다.
>> print(mn, mx)
1 3
```




````python
def numbers(*args):
  num=tuple(args)
  num=list(num)
  mn=min(num)
  mx=max(num)
  l =len(num)
  print("총 {0}개의 숫자가 들어왔습니다. 최소값 {1}, 최대값{2}입니다.".format(l,mn,mx))
  return mn,mx
mn, mx = numbers(1,3,2,1,1)
print(mn,mx)
````

####**예제 10**

길동이는 아래와 같이 코드를 작성하였으나 실행하자 에러 메시지가 떴습니다. 

```
def my_func():
  aa = 1
  bb = 2
  cc = 3
  return aa+bb+cc

o = my_func()
print(o,aa,bb,cc)
```




**10-1**

위 코드의 문제를 설명하세요

* 답: aa,bb,cc는 지역변수로 함수안에서만 적용된다
함수 밖에서도 사용하려면 global을 써서 전역변수로 만들어야 print가능

**10-2**

에러가 뜨지 않도록 함수 my_func()를 수정하세요

````python
def my_func():
  global aa,bb,cc
  aa = 1
  bb = 2
  cc = 3
  return aa+bb+cc

o = my_func()
print(o,aa,bb,cc)
````
