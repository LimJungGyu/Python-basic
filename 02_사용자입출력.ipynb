## **개념정리**
---


사용자에게서 변수를 얻어서 그에 맞는 프로그램을 짜고 싶다면 입력기능이 필요합니다. 또한 원하는 기능을 실행 후 결과물을 확인하고 싶다면 출력기능이 필요합니다. 이번시간에서는 파이썬의 입출력기능을 알아보도록 합니다.







### **(1) 입력**




**1. input()**

입력을 받을 때는 input() 함수를 이용합니다. 원하는 값을 입력할 수 있습니다.

```
my_input = input()
```


````python
my_input = input()
````

**2. 설명추가**

입력을 받을 때 설명을 추가해주고 싶다면 아래와 같이 함수의 괄호 안에 문자열을 삽입합니다.
```
my_input = input("나이:")
```


````python
my_input = input("나이를 입력하세요: ")
````

**3. input()의 자료형**

input() 함수로 바인딩 된 값은 항상 문자열(str)임을 주의하세요

```
my_input = input("나이:")
type(my_input)
```


````python
my_input = input("나이:")
type(my_input)
````

### **(2) 출력**




**1. 아래와 같이 괄호 안의 값을 출력할 수 있습니다.**



```
print(2)
print("hello")
```

````python
print(2)
print("hello")
````

**2. 작은따음표를 아래처럼 출력할 수 있습니다.**

```
print("나는 생각햇다. 'Hello world' ")
```

````python
print("나는 생각했다. 'hello world'")
````

**3. 큰따음표를 아래처럼 출력할 수 있습니다.**

```
print('나는 말했다. "Hello world" ')
```

````python
print('나는 말했다. "Hello world" ')
````

**4. \t를 이용해 탭을 적용합니다.**

```
print("Hello\tworld")
```


````python
print("Hello \t world")
````

**5. \n를 이용해 줄바꿈을 적용합니다.**

```
print("Hello\nworld')
```

````python
print("Hello \n world")
````

**6. 쉼표(,)**

여러개의 인풋을 함께 쉼표(,)를 이용해 출력할 수 있습니다.

```
count=3
color="red"
print(count, color, "apples")
```


````python
count = 3
clolr = "red"
print(count,color,"apples")
````

**7. format()**

format 형식을 이용하여 변수를 정리할 수 있습니다.

```
count=3
color="red"
print("{0} {1} apples".format(count, color))
```


````python
count=3
color="red"
print("{0} {1} apples".format(count, color))
````

**8. %s, %d**

%를 이용하여 변수를 정리할 수 있습니다. 단 숫자는 %d, 문자는 %s를 사용합니다.

```
count=3
color="red"
print("%d %s apples"%(count, color))
```


````python
count=3
color="red"
print("%d %s apples "%(count,color))
````

**9.  sep 옵션**

```
print("Hello", "world", 1, 2, 3, sep='*')
```


````python
print("Hello", "world", 1, 2, 3, sep='*')
````

**10.  end 옵션**

문장이 끝날 때 어떤 문자로 끝날 지 지정해 줍니다. 기본값은 \n (줄바꿈) 입니다.

```
print("Hello", "world", 1, 2, 3, end='*')
```


````python
print("Hello", "world", 1, 2, 3, end='*')
````

#####**예제 14**

아래와 같이 두개의 값을 받아서 더하고 뺀 값을 알려주는 코드를 작성하세요

```
>> 첫번째 숫자를 입력하세요: 
2
>> 두번째 숫자를 입력하세요:
3
두개의 숫자를 더한 값은 5 이고 뺀값은 -1 입니다.
```



````python
first = input("첫번째 숫자를 입력하세요 :")
second = input("두번째 숫자를 입력하세요 :")
s= int(first)+int(second)
m=int(first)-int(second)
print("두개의 숫자를 더한 값은 {0} 이고 뺀값은 {1}입니다".format(s,m))
````
