## 개념정리
---

딕셔너리(dictionary) 는 key와 value로 이루어져 있습니다. 마치 사전처럼 특정 key에 접근해서 특정 key에 해당하는 value를 확인할 수 있습니다. 따라서 딕서녀리에서의 인덱싱 방법은 리스트나 튜플과 다르다는 점을 유의해야 합니다.



### **(1) 생성**

{ } 기호를 통해 딕셔너리를 생성할 수 있습니다.

```
alphabet_dict = { "a":1, "b":2 }
print(alphabet_dict)
type(alphabet_dict)
```




````python
alphabet_dict = { "a":1, "b":2 }
print(alphabet_dict)
type(alphabet_dict)
````

### **(2) 딕셔너리 요소**

다른 군집자료형과 차별되는 딕셔너리만의 특징이 바로 key와 value 입니다. 딕셔너리의 요소는 항상 key-value 한쌍이 존재해야 합니다. **:** 부호를 기준으로 앞에요소가 "key",  뒤에 요소가 "value" 입니다.


####**1. keys()**

이 함수는 딕셔너리의 key 값들만 반환합니다. 

```
alphabet_dict = { "a":1, "b":2 }
print(alphabet_dict.keys())
```
이 때 반환되는 객체는 dict_keys라는 객체임을 확인 할 수 있습니다. 자주 쓰이는 list나 tuple로 변환하여 사용하고 싶으면 타입변환을 해주면 됩니다.

```
alphabet_keys = list(alphabet.keys())
```



````python
alphabet_dict = { "a":1, "b":2 }
print(alphabet_dict.keys())

alphabet_keys = alphabet_dict.keys()
print(alphabet_keys)
alphabet_keys = list(alphabet_keys)
print(alphabet_keys)
````

####**2. values()**

이 함수는 딕셔너리의 value 값들만 반환합니다. 

```
alphabet = { "a":1, "b":2 }
print(alphabet.values())
```
이 또한 타입변환이 가능합니다.

```
alphabet_values = list(alphabet.values())
```



````python
alphabet = { "a":1, "b":2 }
print(alphabet.values())

alphabet_values = alphabet.values()
alphabet_values = list(alphabet.values())
print(alphabet_values)
````

### **(3) 인덱싱과 슬라이싱**

리스트나 튜플과 달리 딕셔너리는 인덱싱만 가능합니다. 인덱싱도 순번으로 인덱싱 하는게 아니라 key 값들로만 인덱싱이 가능합니다.

```
alphabet_dict = { "a":1, "b":2 }
print(alphabet_dict['a'])
```




````python
alphabet_dict = { "a":1, "b":2 }
print(alphabet_dict['a'])
````

순번으로 인덱싱 하면 에러메시지가 뜹니다.

```
alphabet_dict = { "a":1, "b":2 }
print(alphabet_dict[1])
```




````python
alphabet_dict = { "a":1, "b":2 }
print(alphabet_dict[1])  #순번 안됨,범위안됨,
````

### **(4) 기타메서드**

####**1. get()**

이 함수는 딕셔너리의 key값에 해당하는 value 값을 반환합니다. 하지만 인덱싱으로도 value 값을 반환할 수 있기 때문에 사용자 편의에 따른 코드를 사용하면 됩니다.

```
alphabet = { "a":1, "b":2 }
print(alphabet.get("a"))
```


````python
alphabet = { "a":1, "b":2 }
print(alphabet.get("a"),alphabet['a'])  #get함수와 키를이용한 인덱싱은 결과는 같다
````

####**2. items()**

이 함수는 딕셔너리의 key와 value 를 튜플 형태 쌍으로 반환을 합니다.

```
alphabet = { "a":1, "b":2 }
print(alphabet.items())
```


````python
alphabet = { "a":1, "b":2 }
print(alphabet.items())
````

####**3. clear()**

딕셔너리를 초기화합니다. (모든 요소를 제거합니다.)

```
alphabet = { "a":1, "b":2 }
alphabet.clear()
print(alphabet)
```


````python
alphabet = { "a":1, "b":2 }
alphabet.clear()        #다 지워버리는거
print(alphabet)    
````

### **(5) 딕셔너리의 수정과 삭제**

딕셔너리의 인덱싱만 이해하면 수정과 삭제는 리스트나 튜플처럼 자유롭게 가능합니다.

####**1. 수정**

```
alphabet = { "a":1, "b":2 }
alphabet["a"] = 3
print(alphabet)
```




````python
alphabet = { "a":1, "b":2 }
alphabet["a"] = 3
print(alphabet)


#######################삽입##################################
alphabet["c"] = 3
print(alphabet)

######################
n=input("입력:")
alphabet = {'a':0,'b':0,'c':0}
alphabet[n] +=1      #<<----- 덧셈
print(alphabet)
````

####**2. 삭제**

```
alphabet = { "a":1, "b":2 }
del alphabet["a"]
print(alphabet)
```




````python
alphabet = { "a":1, "b":2 }
del alphabet["a"]
print(alphabet)
````

**01-4**

|latte|1200|5

````python
coffee = {"americano":[1000,8],"latte":[1200,5],"milk tea":[1800,4]}
print(coffee,type(coffee))
````

**03-1**

**03-2**

**03-4** 

key 값으로만 구성된 리스트를 생성하고 menu 로 바인딩하세요

````python
menu = list(coffee.keys())
print(menu,type(menu))
````
