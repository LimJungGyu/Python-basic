### **(3) 여러가지 문자열 메소드**

아래 표와 같이 문자열 메소드는 아주 다양합니다. 여기서는 자주 쓰이는 몇가지 함수만 간단히 배워보도록 합니다. 만약 코드를 작성하면서 더 다양한 문자열 메소드가 필요할 경우엔 표를 참고하고 해당 함수를 검색해보고 사용방법을 익혀가는 것도 좋습니다.

|종류|함수|
|--|--|
|string calculation| len(), min(), max(), count()|
|encoding/decoding| encode(), decode()|
|string search|startswith(), endswith(),find(), rfind(), index(), rindex()|
|number/character|isalnum(), isalpha(), isdigit(), isnumeric(), isdecimal()|
|lower/upper| islower(), isupper(), lower(), upper(), swapcase(), istitle(), title(), capitalize()|
|space/strip| isstrip(), rstrip(), strip(), isspace(), center()|
|split/ join / fill | split(), splitlines(), replace(), join(), zfill(), ljust(), rjust()|

#####**1. String calculation**

* len(): 공백까지 포함하여 해당 문자열의 길이 값을 반환합니다.
* max(): 최대값을 구합니다.만약 문자로 이루어졌다면 알파벳 순으로 값을 반환합니다.
* min(): 최소값을 구합니다.만약 문자로 이루어졌다면 알파벳 순으로 값을 반환합니다.

```
a = 'abcd'
b = '1234'
c = 'abcd1234'
print(max(a), min(a), len(a))
print(max(b), min(b), len(b))
print(max(c), min(c), len(b))
```

````python
a= "abc"
b='1234'
c='abcd1234'
d="abAc"
print(max(a), min(a), len(a))
print(max(b), min(b), len(b))
print(max(c), min(c), len(c))  # 숫자가 문자보다 값이 작다
print(max(d), min(d), len(d))  #문자에서는 최소값 a 최대값은 z 소문자보다 대문자가 더 값이 작다
````

#####**2. String search**
* startswith(): 특정 문자로 시작되는지 여부를 확인합니다.
* endswith(): 특정 문자로 끝나는지 여부를 확인합니다.

```
a = 'aBcD'
print(a.startswith('a'))
print(a.endswith('a'))
```


````python
a = 'aBCD.py'
print(a.endswith('py'))
````

#####**3. Lower/upper**

* lower(): 문자열의 모든 알파벳을 소문자로 변환합니다.
* upper(): 문자열의 모든 알파벳을 대문자로 변환합니다.
* capitalize(): 문자열의 첫글자를 대문자로 변환합니다.

```
a = 'aBcD'
print(a.upper())
print(a.lower())
print(a.capitalize())
```


````python
a = 'aBcD'
print(a.upper())
print(a.lower())
print(a.capitalize())
````

#####**4. split/join**

* split() : 문자열의 특정 문자를 기준으로 분할하여 리스트 형태로 반환합니다.
```
a = 'Hi! Happy new year!'
print(a.split())
print(a.split(sep='!'))
```

* join(): 리스트 안의 문자들을 문자열로 변환합니다.
```
print(''.join(a))
print('_'.join(a))
```


````python
a = 'Hi! Happy new year!'
print(a.split())   #디폴트값은 스페이스 바 기준으로 나눠서 리스트에 넣는다
print(a.split(sep='!')) #!를 기준으로
b=a.split()
print(''.join(b))
print(' '.join(b))
print('_'.join(b))
print('*'.join(b))
````

### **(4) 문자열의 연산**

문자열에서도 수치자료형 처럼 몇가지 연산이 가능합니다. 아래 코드를 따라하면서 가능한 연산이 무엇인지 알아보도록 합니다.


#####**1. 덧셈**

문자열끼리는 덧셈이 가능합니다. 단 문자열이 아닌 수치자료형이나 리스트자료형처럼 다른 자료형과는 덧셈이 불가 합니다.

```
a = 'abc'
b = 'cde'
c = 2
print(a+b)
print(a+c)
```


````python
a = 'abc'
b = 'cde'
c = 2
print(a+b)   #abccde
print(a+c)    #문자열에서의 덧셈은 문자열끼리만 가능하다 오류남
````

#####**2. 곱셈**

문자열과 int 값은 곱셈이 가능합니다. 다른 자료형이나 수치 자료형이라도 int가 아닌 float 값 등은 곱셈이 불가능합니다.

```
a = 'abc'
b = 3
c = 3.0
print(a*b)
print(a*c)
```


````python
a = 'abc'
b = 3
c = 3.0
print(a*b)  #문자열의 곱셈은 정수값이랑만 가능하고 정수값은 몇번 반복할것인가를 결정한
print(a*c)  #float값이랑은 곱셈이 불가하다 오류남
````
