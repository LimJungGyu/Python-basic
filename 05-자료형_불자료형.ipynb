### **(2) 비교연산자**

위에서 우리는 수치자료형의 두 값을 비교하였습니다. 이와 같이 비교연산자를 이용해서 두 값을 비교할 수 있습니다. 또한 두 값을 비교한 결과는 True와 False로 나오며 이 때 타입은 불 자료형(bool)이 됩니다.


#####**1. 같다**

```
a = 10
b = 10
c = 20
print(a==b)
print(a==c)
```




````python
a = 10
b = 10
c = 20
print(a==b)
print(a==c)
````

#####**2. 다르다**

```
a = 10
b = 10
c = 20
print(a!=b)
print(a!=c)
```




````python
a = 10
b = 10
c = 20
print(a!=b)
print(a!=c)
````

#####**3. 크다/작다**

````python
a = 10
b = 10
c = 20
print(a>b)
print(a<c)
print(a>=b)
print(a<=c)
````

### **(3) Identity 연산자**

이번에는 두 변수가 같은 메모리에 있는지 확인하는 연산자를 배워 볼 것입니다. 


#####**1. is** 

is 연산자는 두 변수가 같은 메몰를 가리키는지 알려주는 연산자입니다.

```
a = 10
b = 10
c = 10.0

print(a==b)
print(a is b)

print(a==c)
print(a is c)

```




````python
a = 10
b = 10
c = 10.0

print(a==b)
print(a is b)

print(a==c)      ##10과 10.0은 같다
print(a is c)    #하지만 10과 10.0은 정확히는 타입이 다르고 컴퓨터상으로 다르게 인식되기때문에 정확히 알려줌
````

#####**2. is not**

is not 연산자는 두 변수가 다른 메모리를 가리키는지 알려주는 연산자입니다.
```
a = 10
b = 10
c = 10.0

print(a!=b)
print(a is not b)

print(a!=c)
print(a is not c)

```




````python
a = 10
b = 10
c = 10.0

print(a!=b)
print(a is not b)

print(a!=c)    #is랑 정확히 반대 개념
print(a is not c)
````

### **(4) 논리연산자**
---
논리 연산자에는 and, or, not이 있고 각각의 진리표 (Truth Table)은 다음과 같습니다.




#####**1. AND** 
모두 True여야 True


|A|B| 결과 |
|--|--|--|
|False|False|False| 
|False|True|False| 
|True|False|False| 
|True|True|True| 



```
a = True
b = False
print(a and b)
```




````python
a=10
b=20
c=30
print(a<b and b<c and a>c) #둘중에 하나라도 false이면 false이다
````

#####**2. OR**
모두 False 일때만 False

|A|B| 결과 |
|--|--|--|
|False|False|False| 
|False|True|True| 
|True|False|True| 
|True|True|True| 

```
a = True
b = False
print(a or b)
```

````python
a = True
b = False
print(a or b)
````

#####**3. NOT**
결과 반대

|A|결과 |
|--|--|
|False|True| 
|False|False| 


```
a = True
b = False
print(not a)
print(not b)
```

````python
a = True
b = False
print(not a)
print(not b)
````

### **(5) 자료형의 True/False**
---

|값	|자료형|	True OR False |
|--|--|--|
"Hello, World!"|	str	|True
""	|str	|False
[1,2,3,4]	|list|	True
[]|	list	|False
(1,2,3,4)	|tuple|	True
()|	tuple|	False
{}|	dictionary|	False
1	|int|	True
2	|int|	True
0	|int|	False

#####**1. 문자**

문자가 하나라도 포함되어 있으면 **True** 입니다.

```
print(bool("abc"))
print(bool(""))
```

````python
print(bool("abc"))
print(bool(1))
print(bool(""))  string 이든 int 든 무엇이라도 들어가있으면
````

#####**2. 리스트/튜플/딕셔너리**

리스트(튜플/딕셔너리)에 하나라도 포함되어 있으면 **True** 입니다

```
print(bool([1,2]))
print(bool([]))
```

````python
print(bool([1,2]))
print(bool([])) 
````

#####**3. 수치자료형**

0일때만 **False** 입니다.

```
print(bool(10))
print(bool(-10))
print(bool(0))
```

````python
print(bool(10))
print(bool(10.23))
print(bool(-10))
print(bool(0))
print(bool(0.0))  #0을 제외한 모든 수치는 true
````
