## 개념정리
---

또 다른 반복문인 whlie문은 for문과 동일하게 특정 코드를 반복해서 진행합니다. 그렇지만 for문과 달리 반복문 내에 반복되는 특정 변수가 없으며 필요시 직접 입력해주어야 합니다. 

### **(1) 문법**













#### **1. 기본구조**

while문의 기본 구조는 아래와 같습니다.



```
while <조건문>:
  <반복되는 코드>
  ...
```

조건문은 항상 bool 형태, 즉 True 혹은 False임을 유의하세요. while은 조건문이 True인 동안 계속해서 반복됩니다. <조건문> 자리에 a가 5보다 작다는 조건을 입력해봅시다. 반복문이 시작할 때 a가 0이기 때문에 a<5라는 조건은 **True** 가 됩니다. 조건문이 **True**인 동안 a는 1씩 증가하다가 a가 5 이상이 되는 순간 조건문은 **False**가 되며 반복문이 종료됩니다.

```
a=0
while a<5:
  print(a)
  a+=1
```

````python
a=0
while a<5:  #True 인 상태이기 때문에 while문이 시작됨
  a+=1  
  print(a)   #a가 4에서 시작해서 while문 안에서 5가 되어도 이미 시작된 루프는 돌아간후 끝난다
print("end")
````

####**2. 무한루프**

이번에는 <조건문> 자리에 직접 **True** 를 입력하겠습니다. 이렇게 되면 반복문의 조건문이 항상 **True인** 상태기 때문에 다른 조건을 코드 아래에 추가해 주지 않으면 무한루프가 됩니다. (아래 코드는 무한 루프이기 때문에 굳이 실행해 보지 않아도 됩니다.)

```
a=0
while True:
  print(a)
  a+=1
```

###**(3) break과 continue**



####**1. break**
 따라서 무한루프의 경우 보통 반복문 내에 종료코드를 넣어줍니다. 따라서 아래 코드 예시와 같이 특정조건이 될 때 종료하는 코드를 작성합니다. 종료는 앞서 배운 **break** 을 이용합니다.

 ```
a=0
while True:
  print(a)
  a+=1
  if a==5:
    break
```

````python
a=0
while True:
  print(a)
  a+=1
  if a==5:
    break
````

####**2. continue**

만일 특정 조건에서 루프를 끝내지 않고 다음 루프로 넘어가고 싶다면 **continue**를 사용합니다.

 ```
a = 0
while True:
  print(a)
  a+=1
  if a==5: 
    continue
  if a==10:
    break
```

````python
a = 0
while True:
  print(a)
  a+=1
  if a==5: 
    continue
  if a==10:
    break
````

####**예제 04**

while문을 이용해 아래 결과를 내는 함수 **bomb()** 를 작성하세요. 음수가 입력될 경우 다른 에러메시지를 출력하세요.


```
>> bomb() 
Enter timer: 6
befor 6 sec.
befor 5 sec.
befor 4 sec.
befor 3 sec.
befor 2 sec.
befor 1 sec.
BOMB!
>> bomb()
Enter timer: -3

```

````python
def bomb():
  timer = int(input("Enter timer : "))
  while timer != 0:
    if timer>0:
      print("before {0} sec".format(timer))
      timer -=1
    elif timer<0:
      print("TIMER ERROR")
      break
    print("BOMB!")
bomb()
````

####**예제 06**
* while 문을 이용해 1부터 100까지 값을 출력합니다.
* 만약 출력값이 2의 배수면 mul2, 3의 배수면 mul3, 5의 배수면 mul5에 넣으세요 
* mul2, mul3, mul5는 리스트이며 중복가능하게 각 리스트에 숫자를 넣습니다.

````python
j=0
mul2 =[]
mul3 =[]
mul5 =[]
while j<=5:
  j+=1
  i=1
  while i>=1 and i<=100:
    if j==1 and i%2==0:
      mul2.append(i)
    if j==2 and i%3==0:
      mul3.append(i)
    if j==3 and i%5==0:
      mul5.append(i)
    i+=1
print(mul2)
print(mul3)
print(mul5)
````

####**예제 07** 

출처: 백준 https://www.acmicpc.net/problem/1110

0보다 크거나 같고, 99보다 작거나 같은 정수가 주어질 때 다음과 같은 연산을 할 수 있다. 먼저 주어진 수가 10보다 작다면 앞에 0을 붙여 두 자리 수로 만들고, 각 자리의 숫자를 더한다. 그 다음, 주어진 수의 가장 오른쪽 자리 수와 앞에서 구한 합의 가장 오른쪽 자리 수를 이어 붙이면 새로운 수를 만들 수 있다. 다음 예를 보자.

26부터 시작한다. 2+6 = 8이다. 새로운 수는 68이다. 6+8 = 14이다. 새로운 수는 84이다. 8+4 = 12이다. 새로운 수는 42이다. 4+2 = 6이다. 새로운 수는 26이다.

위의 예는 4번만에 원래 수로 돌아올 수 있다. 따라서 26의 사이클의 길이는 4이다.

N이 주어졌을 때, N의 사이클의 길이를 구하는 함수 **find_cycle()**을 작성하시오.



```
>>find_cycle()
N을 입력하세요(0부터99까지): 26
cycle은 4 입니다.
>>find_cycle()
N을 입력하세요(0부터99까지): 55
cycle은 3 입니다.
>>find_cycle()
N을 입력하세요(0부터99까지): 0
cycle은 1 입니다.
>>find_cycle()
N을 입력하세요(0부터99까지): 1
cycle은 60 입니다.

```



````python
def find_cycle():
  number=input("N을 입력하세요(0부터99까지): ")
  n = list(number)
  n[0]=int(n[0])
  n[1]=int(n[1])
  i=1
  p=0

  m=["",""]
  m[0]=n[1]
  m[1]=n[1]+n[0]
  print(m)
  while i<10:
    if m[0] == n[0] and m[1]==n[1]:
      break
    else:
      if m[1]>10:
        p=m[0]
        m[0]=m[1]-10
        m[1]= m[1]+p-10
      else:
        p=m[0]
        m[0]=m[1]
        m[1]= m[1]+p
      print(m[0],m[1])
      i=i+1
  print("cycle은 {0}".format(i))    
find_cycle()
````
