## 개념정리1 : time
---

시간과 관련된 프로그램은 아주 많습니다. 이를 위해 유용한 라이브러리인 time 모듈을 알아봅니다. time 모듈에는 함수가 굉장히 많습니다. 그 중 자주 사용되는 함수 몇가지 위주로 공부해봅니다.


### **(1) 라이브러리 임포트(import)**

아래와 같이 **time** 모듈을 import 해줍니다.

```
import time
```


````python
import time
````

### **(2) 시간확인**



####**1. time.time**

time.time()은 UTC(Universal Time Coordinated 협정 세계 표준시)를 사용하여 현재 시간을 실수 형태로 돌려주는 함수이다. 1970년 1월 1일 0시 0분 0초를 기준으로 지난 시간을 초 단위로 돌려준다.

```
time.time()
```



````python
time.time()
````

####**2. time.ctime**
현재 시간을 알아보기 쉬운 문자열 형식으로 반환합니다.

```
t = time.ctime()
print(t)
print(type(t))

```



````python
t = time.ctime()
print(t)
print(type(t))
````

### **(3) 시간조절**



####**1. time.sleep**

time.sleep은 코드를 일정 시간동안 멈출 수 있습니다. 주로 반복문 내에서 시간 간격을 두고 실행하고 싶을 때 쓰입니다.

```
for i in range(10):
    print(i)
    time.sleep(1)
```



````python
for i in range(10):
    print(i)
    time.sleep(1.5)  ##sleep안에는 초단위 숫자가 들어간다
````

## 개념정리2 : calendar
---

달력이 필요할 때 파이썬에서 주로 쓰는 모듈입니다. 날짜와 관련된 여러가지 편리한 함수를 사용할 수 있습니다.

### **(1) 라이브러리 임포트(import)**

아래와 같이 **time** 모듈을 import 해줍니다.

```
import calendar
```


````python
import calendar
````

### **(2) 달력 보기**



####**1. calendar.prmonth**

calendar.prmonth는 연도와 월을 입력하면 해당 월의 달력을 출력해줍니다.

```
calendar.prmonth(연도, 월)
```



````python
calendar.prmonth(2015, 12)
````

### **(3) 날짜 반환**



####**1. calendar.weekday**

weekday 함수는 연도, 월, 일을 입력하면 0에서 6까지의 정수 반환합니다. 각 숫자는 월요일 부터 일요일을 의미합니다. 

```
정수 = calendar.weekday(연도, 월, 일)
# 정수값은 0~6 이며, 월~일을 의미
```



````python
week_dic = {0:"월",4:"금"}
week_int = calendar.weekday(2021, 12, 10)
week_dic[week_int]
````

####**2. calendar.monthrange()**

calendar.monthrange는 연도와 월을 입력하면 이달이 무슨 요일인지,그리고 며칠까지 있는지를 튜플 형태로 출력합니다. 

```
첫번째요일, 일수 = calendar.monthrange(연도, 월)
```



````python
int_fir,int_days = calendar.monthrange(2021,12)
int_fir,int_days
````

타이머를 만듭시다. 이 타이머는 10초 단위로 알림을 하고 10 부터는 카운트 합니다. (실제로 초단위의 타이머를 만듭니다.)

````python
def 타이머(t):

  while t != 0:
    if t>10: 
       print(t)
       t-=10    
    else:
      print(str(t)+"!")
      t-=1
    time.sleep(1)
  print("삐빅")
타이머(60)
````

**예제 02**

시간을 알려주는 챗봇을 만들어 봅시다. 사용자가 입력한 문장에서 다음과 같은 문장이 있으면 그에 따른 대답을 합니다.

단어|대답|
--|--
몇시| 시간 출력
몇분| 분 출력
무슨요일| 요일 출력
몇월| 월 출력
며칠or 몇일 | 일 출력
그만 | 종료

```
>> 시리()
>> 질문해주세요: 지금몇시니?
지금은 8시 입니다.
>> 질문해주세요: 오늘 무슨요일이야
오늘은 금요일 입니다.
>> 질문해주세요: 그만
종료되었습니다.
```


>>



````python
def 시리():
  t=time.time()
  a_list=list(time.gmtime(t))
  week_dic={0:"월",1:"화",2:"수",3:"목",4:"금",5:"토",6:"일"}
  while True:
    user = input("질문해주세요:")
    if user == "그만": break
    elif "몇시" in user:
      print("지금은 {0}시 입니다".format(a_list[3]))
    elif "몇분" in user:
     print("지금은 {0}분 입니다".format(a_list[4]))
    elif "무슨요일" in user:
      print("오늘은 {0}요일 입니다".format(week_dic[a_list[6]]))
    elif "몇월" in user:
     print("지금은 {0}월 입니다".format(a_list[1]))
    elif "몇일" in user or "며칠" in user:
     print("지금은 {0}일 입니다".format(a_list[2])) 
시리()
````

**예제 03**

2000년부터 2020년까지 각 해의 날짜의 수를 구하세요

ex) 2020년의 날짜는 총 365일 입니다.


````python
for j in range(21):
  year_day=0
  for i in range(12):
    a, b = calendar.monthrange(2000+j, i+1)
    year_day += b
  print("{0}년의 날짜는 총{1}일입니다.".format( 2000+j, year_day ))
````
