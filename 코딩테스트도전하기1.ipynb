### 2. 두개뽑아서 더하기
---
주소: https://programmers.co.kr/learn/courses/30/lessons/68644


**나의 풀이**

```python
def solution(numbers):
    answer = []
    for i in range(len(numbers)):
        for j in range(len(numbers)):
            if i!=j:
              answer.insert(j,numbers[i]+numbers[j])
              continue
    answer=set(answer)
    answer=list(answer)
    answer.sort()    
    return answer
solution([2,1,3,4,1])
```

**다른사람의 풀이**

```python
def solution(numbers):
    answer = []
    for i in range(len(numbers)):
        for j in range(i+1, len(numbers)):
            answer.append(numbers[i] + numbers[j])
    return sorted(list(set(answer)))
```

### 3. 음양더하기
---
주소: https://programmers.co.kr/learn/courses/30/lessons/76501?language=python3


**나의 풀이**

```python
def solution(absolutes, signs):
    result=0
    for i in range(len(signs)):
        if signs[i] == True:
            result+=absolutes[i]
        else:
            result-=absolutes[i]
    answer = result
    return answer
```

**다른사람의 풀이**

```python
def solution(absolutes, signs):
    return sum(absolutes if sign else -absolutes for absolutes, sign in zip(absolutes, signs))
```

**나의 풀이**

```python
def solution(s):
  answer = ''
  a=list(s)
  if len(a)%2==0:
    answer = ''.join(a[len(a)//2-1]+a[len(a)//2])
  else:
    answer= str(a[len(a)//2])
  return answer
solution("qwer")
```
