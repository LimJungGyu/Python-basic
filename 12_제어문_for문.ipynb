#### **1. 군집자료형과 쓰는 for문**

군집자료형의 요소가 차례대로 i 에 바인딩 됩니다. for문 아래에서 i가 변수로 작동됩니다. 아래 예시에서 i는 모두 1,2,3 순으로 출력됩니다.  변수가 필요 없으면  세번째 용례처럼 언더바(_)로 표기하기도 합니다.

```
for i in (1,2,3):
  print(i)

for i in [1,2,3]:
  print("반복", i)

for _ in "123":
  print("반복")

```

````python
for i in (1,2,3):
  print(i)
  print("---루프내 출력---")
print("<<<루프 외 출력>>>")
````

````python
for i in [1,2,3]:
  print("반복", i)
````

````python
for _ in "123":
  print("반복")
````

####**2. range와 함께 쓰는 for문**

굳이 군집자료형을 만들어 놓지 않아도 반복문을 쓸 수 있습니다. 이 때는 range를 이용하는데, range도 슬라이싱 과 마찬가지로 처음, 끝, 스텝이 존재합니다. 

* range(n): 변수 i는 0부터 n-1 까지 총 n 회 반복됩니다.

```
for i in range(3):
  print(i)
```

````python
for i in range(3):
  print(i+1)
````

* range(n, m): 변수 i는 n부터 m-1까지 총 m-n회 반복됩니다.

```
for i in range(3, 10):
  print(i)
```

````python
for i in range(3, 10):
  print(i)
````

* range(n, m, s): i는 n부터 m-1까지 s씩 건너뛰며 반복됩니다.

```
for i in range(3, 10, 2):
  print(i)
```

````python
for i in range(3, 10, 2):
  print(i)
````

#### **1. continue**  

for 문 안에서 continue를 만나면 즉시 for문의 처음으로 돌아가고 다음 반복이 진행됩니다.

```
for i in range(100):
  if i<80:
    continue  
  i_add = i+100
  print(i_add)
```

```
for i in range(100):
  if i>=80:
    i_add = i+100
    print(i_add)
  ```

코드 예시의 위아래의 결과는 동일합니다. 하지만 첫번째 코드에서는 80미만 숫자들이 continue에서 바로 루프를 종료하고 다음 루프로 넘아가는 반면 두번째 코드에서는 80미만 숫자들이라도 if 문 끝까지 진행이 됩니다. 예시처럼 간단한 코드의 경우는 별 차이가 없을지라도 코드가 복잡해질수록 계산량이 차이가 납니다. 효율적인 코드 작성을 위해서는 continue를 적절히 사용해 줄 필요가 있습니다. 


````python
for i in range(100):
  if i<80:
    continue  
  i_add = i+100
  print(i_add)
````

#### **2. break**  

for 문 안에서 break 을 만나면 즉시 for문이 종료됩니다. (모든 루프가 돌아가지 않았어도)

```
for i in range(100):
  if (i%5==0) and (i>=5) and (i%2==0):
    print(i)
    break
```

원하는 결과를 얻어서 더이상 반복문을 진행할 필요가 없을 때 보통 쓰입니다. 위 코드로 가장 5와2의 최소공배수를 찾을 수 있습니다.


````python
for i in range(100):
  if (i%5==0) and (i>=5) and (i%2==0):
    print(i)
    break
````


### **(4) for문을 더 효율적으로**


#### **1. zip**  

두개의 sequnce를 차례대로 같이 나열하고 싶을 때 쓰입니다.

```
a_list = ["A","B","C"]
b_list = ["a", "b", "c"]
for a, b in zip(a_list,b_list):
  print("a: %s, b: %s"%(a, b))
```



````python
a_list = ["A","B","C"]
b_list = ["a", "b", "c"]
for a, b in zip(a_list,b_list):
  print("a: %s, b: %s"%(a, b))

for i in range(3):
  print(a_list[i],b_list[i])
````

#### **2.enumerate**  

sequence를 나열할 때 인덱스를 함께 나열하고 싶을 때 쓰입니다. 

```
a_list = ["A","B","C"]
for idx, a in enumerate(a_list):
  print( "idx: %i, a: %s"%(idx, a) )
```



````python
a_list = ["A","B","C"]
for idx, a in enumerate(a_list):
  print( "idx: %i, a: %s"%(idx, a) )
````


### **(5) 다중 for문**
for문 아래 for문을 또 추가 할 수 있습니다. 다만 다중 for문일 때 코드의 작동순서를 잘 파악할 필요가 있습니다. 삼중, 사중 for문도 구현은 가능합니다. 하지만 for문이 추가될때마다 계산량이 급격히 증가하기 때문에 되도록이면 이중 for문을 넘어가지 않는것이 좋습니다.

```
for i in range(3):
  print("i:", i)
  for j in range(3):
    print("j:",j)
```



````python
for i in range(3):
  print("i:", i)
  for j in range(3):
    print("j:",j)
````

코드가 아직 익숙하지 않은 상태에서 다중 for문을 보면 해석하기 어려울 수 있습니다. 이것저것 변수를 건드려보고, 출력을 바꾸면서 최대한 익숙해지도록 하는 것이 좋습니다. 아래 코드가 어떻게 나올지 예상해보고 출력해 봅시다.

```
for i in range(10):
    print(i, end=": ")
    for j in range(1, 11):        
        print(j, end=" ")
    print()
```



````python
for i in range(10):
    print(i, end=": ")
    for j in range(1, 11):        
        print(j, end=" ")
    print()
````

```
>> add_nums(1)

>> add_nums(2)
```

````python
def add_nums(n):
  sum=0
  for i in range(1,n+1):
    sum = sum+i
  print(sum)
add_nums(1)
add_nums(2)
add_nums(10)
````

####**예제 08**

학생들의 번호와 점수가 딕셔너리에 아래와 같이 저장되어 있습니다.

```
st_scores = {1:80, 2:55, 3:80, 4:35, 5:75, 6: 80, 7:40, 8: 90, 9:55, 10:20}
```



````python
st_scores = {1:80, 2:55, 3:80, 4:35, 5:75, 6: 80, 7:40, 8: 90, 9:55, 10:20}
````

**08-1** 

전산 실수로 인해 점수가 실제 점수보다 10점이 깍인것을 발견했습니다. 점수를 모두 10점씩 올리세요

````python
for i in st_scores:
  score=st_scores.get(i)
  st_scores[i]=score+10
  continue
print(st_scores)
````

**08-2** 

60점 미만인 학생들에게 "재평가가 실시될 예정입니다.", 90점 이상인 학생들에게 "우수장학금이 지급됩니다."를 출력하는 코드를 작성하세요.  코드내에 continue를 반드시 포함하세요

````python
for i in range(len(st_scores)):
  score = st_scores.get(i+1)
  if  score < 60 :
    print("{0}번 학생은 재평가가 실시될 예정입니다".format(i+1))
  elif score >= 90 :
    print("{0}번 학생은 우수장학금이 지급됩니다".format(i+1))
  else:
    continue
````
