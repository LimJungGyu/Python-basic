## **개념정리**
---

코드를 작성하면서 우리는 다양한 데이터 자료형을 접하게 됩니다. 그 중 수치 자료형은 개수, 가격, 날짜 등 여러가지 데이터를 효과적으로 처리할 수 있습니다. 수치 자료형에 대해 알아보고 데이터 타입이 수치 자료형일 때 어떤 부가적인 연산들을 수행할 수 있는지 알아보도록 합니다. 

### **(1) 수치자료형의 종류**
수치 자료형에는 다음과 같은 종류가 있습니다.


#####**1. 정수형자료형**

정수값을 의미합니다. 타입 함수를 이용하여 데이터 타입을 출력해봅니다.

```
num = 2
print(num, type(num))
num = 0
print(num, type(num))
num = -5
print(num, type(num))
```




````python
num =2
print(num, type(num))
num=0
print(num,type(num))
num=-5
print(num,type(num))
````

#####**2. 실수형 자료형**

실수값을 의미합니다. 타입 함수를 이용하여 데이터 타입을 출력해봅니다.

```
num = 0.2
print(num, type(num))
num = -0.4
print(num, type(num))
```




````python
num=0.2
print(num,type(num))
num = -0.4
print(num,type(num))
num =1/5
print(num,type(num))
````

#####**3. 복소수형 자료형**

복소수는 제곱했을 때 음수가 되는 수입니다. 아래와 같이 정의할 수 있습니다.

```
num = 1+4j
print(num, type(num))
```




````python
num =1+5j
print(num,type(num))
````

복소수 자료형 경우  파이썬 내장 함수를 통해 실수값, 허수값, 컬례 복소수 등을 알 수 있습니다.

```
print(num.real) # 실수값
print(num.imag) # 허수값
print(num.conjugate()) # 컬례복소수
```

````python
num.real, num.imag, num.conjugate()
````

```
>> 첫번째 숫자:

>> 두번째 숫자:
3
10 나누기 3은 몫이 3이고 나머지는 1입니다.
```

````python
fir = input("첫번째 숫자:")
sec = input("두번째 숫자:")
a = int(fir)//int(sec)
b = int(fir)%int(sec)
print("{0} 나누기 {1}는 몪이 {2}이고 나머지는 {3} 입니다.".format(fir,sec,a,b))
````
