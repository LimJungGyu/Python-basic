## 개념정리
---

수학에서 쓰이는 집합의 개념과 비슷하게 파이썬에서도 집합을 사용할 수 있습니다. 


### **(1) 생성**

{ } 괄호를 통해서 집합을 생성 할 수 있습니다. 딕셔너리 또한 { } 괄호를 통해 생성하지만 key와 value를 모두 적어야 하는 반면 집합은 괄호 안에 값들만 들어가는 차이점이 있습니다. 

```
a = {1,2,3}
print(a, type(a))


```




````python
a = {1,2,3}
print(a, type(a))
 
````

### **(2) 집합의 특징**

다른 군집자료형(문자열, 리스트, 튜플, 딕셔너리)과 다른 집합만의 특징에는 두가지가 있습니다.

* 중복을 허용하지 않습니다.
* 순서가 없습니다.



#### **1. 중복불허용**

집합에는 중복되는 값이 존재 하지 않습니다. 따라서 아래와 같이 초기값으로 중복된 요소를 넣더라도 중복값은 자동적으로 제거됩니다. 



```
a = {1,2,2,2,3,1}
print(a)
```



````python
a = {1,2,2,2,3,1}
print(a)
````

#### **2. 순서없음**

집합에는 순서 개념이 없습니다. 따라서 군집 자료형을 집합으로 만드는 순간 자동으로 정렬됩니다. 또한 순서 개념이 없기 때문에 인덱싱이나 슬라이싱이 불가능 합니다.


```
a = set([3,2,2,2,1])
print(a)
print(a[1]) # 에러메시지 발생
```



````python
a = set([3,2,2,2,1])
print(a)
print(a[1]) #순서가 없기 때문에 인덱싱도 되지않는다 에러
````

이러한 두가지 특징을 가지고 있기 때문에 다른 군집 자료형에 대해 중복을 제거하면서 정렬을 해야 할 때 타입을 집합으로 바꾸기도 합니다. 

```
a_list = [1,2,3,2]
a = set(a_list)
a_list = list(a)
print(a_list)
```



````python
a_list = [1,2,3,2]
a = set(a_list)
a_list = list(a)
print(a_list)  #  set자체만으로는  잘 안쓰지만 중복제거용으로 쓴다
````

### **(3) 집합의 연산**

리스트나 튜플과 달리 집합은 + 나 * 연산이 불가능 합니다. 하지만 수학적 집합 개념과 마찬가지로 교집합, 합집합 차집합의 연산을 할 수 있습니다. 

#### **1. 교집합**

합집합은 **&** 기호나 **union()** 을 사용합니다. 참고로 **&**기호는 대부분의 언어에서 **and**를 뜻합니다. 

```
a = set([1,2,3])
b = set([3,4,5])
print(a&b)
print(a.intersection(b))
```



````python
a = set([1,2,3])
b = set([3,4,5])
print(a&b)
print(a.intersection(b))  ##a&b랑 같은 결과가 나온다
````

#### **2. 합집합**

합집합은 **|** 기호나 **union()** 을 사용합니다. 참고로 **|** 기호는 대부분의 언어에서 **or** 를 뜻합니다.

```
a = set([1,2,3])
b = set([3,4,5])
print(a|b)
print(a.union(b))
```



````python
a = set([1,2,3])
b = set([3,4,5])
print(a|b)
print(a.union(b))
````

#### **2. 차집합**

차집합을 구할 때는 산술연산을 사용합니다. **-** 기호를 이용합니다. 또한 **difference()**를 사용할 수도 있습니다.

```
a = set([1,2,3])
b = set([3,4,5])
print(a-b)
print(a.difference(b))
```



````python
a = set([1,2,3,4])
b = set([3,4,5,2])
print(a-b)
print(a.difference(b))
````

### **(4) 기타메서드**

####**1. add()**

리스트의 append() 기능과 유사합니다. 집합에 값을 하나 추가합니다. 다만 append()는 마지막에 값이 추가되는거와 달리 집합에서는 정렬된 위치에 값이 추가됩니다.

```
a ={1,2,3,5}
a.add(4)
print(a)
```


````python
a ={1,2,3,5}
a.add(4)
print(a)
````

####**2. update()**

값을 한번에 여러개 추가해 줄 수 있습니다.

```
a = {1,2,3,5}
a.update([4,0])
print(a)
```


````python
a = {1,2,3,5}
a.update([4,0])
print(a)
````

####**3. remove()**

값 하나를 제거합니다.

```
a = {1,2,3,5}
a.remove(2)
print(a)
```


````python
a = {1,2,3,5}
a.remove(2)
print(a)
````

####**4. 내장함수**

다른 군집 자료형과 마찬가지로 내장함수를 사용할 수 있습니다.

* len(): 길이를 구합니다.
* max(): 최대값을 구합니다.
* min(): 최소값을 구합니다.
* sum(): 모든값을 더합니다.


```
a = {1,2,3,5}
len(a), max(a), min(a), sum(a)
```


````python
a = {1,2,3,5}
len(a), max(a), min(a), sum(a)
````
