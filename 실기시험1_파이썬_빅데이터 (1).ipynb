
* 그림과 같은 2차원 배열을 `arange`와 `reshape`를 이용해서 생성하기 (5점)


````python
arr = np.array([[0,1,2,3,4,5],[10,11,12,13,14,15],[20,21,22,23,24,25],[30,31,32,33,34,35],[40,41,42,43,44,45],[50,51,52,53,54,55]])
print(arr,arr.shape)
````
