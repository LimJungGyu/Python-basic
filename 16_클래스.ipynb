#### **1. __init__**

__init__ 함수를 통해 클래스의 인풋을 정의 할 수 있습다. 인풋 뿐만 아니라 클래스가 시작 할 때 기본 값들을 세팅할 수 있습니다.

```
class Student():
    def __init__(self, name, subject):
      self.name = name
      self.subject = subject
      self.level = '3개월차'

me = Student("이름", "수업")

```

````python
class Student():              #클래스의 인풋은 클래스명옆의 ()의 인풋을 보는게 아니고 init의 인풋을 봐야한다
    def __init__(self, name, subject):  #클래스를 만들면 init함수가 필요하다 self는 클래스 공통으로 쓰이는 변수이고 
      self.name = name                  #클래스내의 모든 변수에 self를 붙여줘야한다
      self.subject = subject
      self.level = '3개월차'

me = Student("이름", "수업")
````

````python
import random
class MyGame():
  def __init__(self, 닉네임, 직업):
    self.닉네임 = 닉네임
    self.직업 = 직업  
    self.체력 = 50 
    self.최대체력 =100
    self.레벨 =1
    self.현재장소 = "집"
    self.골드 = 1000
    self.장소리스트 = ["집", "사냥터"]

    self.메뉴화면() # 함수

  def 메뉴화면(self):
    while True:
      print("현재장소:", self.현재장소, "골드:", self.골드, "체력: ", self.체력, "/", self.최대체력)
      print("|| 1.프로필 || 2.행동 || 3.이동 || 4.게임종료 ||")
      선택 = input(">> 선택하세요: ")
      if 선택=='1':
        self.선택_프로필()
      elif 선택=='2':
        self.선택_행동()
      elif 선택=='3':
        self.선택_이동()     
      elif 선택=='4':
        break
      else:
        print("잘못 입력하셨습니다.")

  def 선택_프로필(self):
    print("**********************")
    print("* 이름: ", self.닉네임)
    print("* 직업: ", self.직업)
    print("**********************")

  def 선택_이동(self):
    print(self.장소리스트)
    이동장소 = input()
    print(">> ", self.현재장소, "에서 ", 이동장소,"로 이동합니다.")
    self.현재장소=이동장소

  def 선택_행동(self):
    if self.현재장소=='집':
      self.장소_집()
    elif self.현재장소=='사냥터':
      self.장소_사냥터()
    else: 
      print(">> 잘못 입력하였습니다.")
  
  def 장소_집(self): 
    선택 = input(">> 행동을 선택하세요: 1.휴식(골드 '15'사용) 2.취침(골드'100'사용) 3.레벨업(골드'1000'사용)")
    if int(선택)==1 and self.골드>=15:
      self.체력=min(self.체력+10, self.최대체력)
      self.골드-=15
      print(">> 골드 (15)사용 체력이 (10)회복 하였습니다.")
    elif int(선택)==2 and self.골드>=100:
      self.체력=min(self.체력+100, self.최대체력)
      self.골드-=100
      print(">> 골드 (100)사용 체력이 (100)회복 하였습니다.")
    elif int(선택)==3 and self.골드>=1000:
      self.레벨+=1
      self.골드-=1000
      print(">> 골드 (1000)사용 레벨이 (1)올랐습니다.")
    else:
      print(">> 잘못 입력하셧 거나 골드가 모자랍니다.")
  def 장소_사냥터(self):
    선택2 = input(">>행동을 선택하세요: 1.사냥 2.채집.3.돌아가기")
    if int(선택2)==1:
      self.사냥()
    elif int(선택2)==2:
      print("골드 10획득")
      self.골드+=10
      print("현재골드는 :",self.골드)
    elif int(선택2)==3:
      self.메뉴화면
    else:
      print(">> 잘못 입력하였습니다.")
  def 사냥(self):
    print("적만남")
    적종류=[['토끼',10,5],['토끼',10,5],['토끼',10,5],['토끼',10,5],['여우',30,20],['여우',30,20],['여우',30,20],['늑대',50,30],['호랑이',100,40]]
    선택3 = input(">>행동을 선택하세요: 1.전투 2.돌아가기")
    if int(선택3) == 1:
      적=적종류[random.randint(0,4)]
      적체력 = 적[1]
      print(적[0],"을/를 만났습니다")
      while True:
        if 적체력 >=0:
          데미지 =random.randint(0,적[2])
          print("적이 {0}의 데미지로 공격합니다".format(데미지))
          선택4 = input(">>>1.공격 2.방어")
          if 선택4 == "1":
            self.체력-=데미지
            적체력-= random.randint(1,10)*self.레벨
            print("나의 체력:{0}///////적의 체력:{1}".format(self.체력,적체력))
          elif 선택4=="2":
            print("피했습니다")
          else:print("잘못입력하셧습니다")
        elif self.체력<=0:
          self.레벨 -= 2
          print("사망하셧습니다")
          break
        elif 적체력<=0:
          self.골드 += 적[1]*random.randint(1,10)
          self.레벨 += 1
          print("사냥 성공!!!!! 현재골드:{0} 현재레벨:{1}".format(self.골드,self.레벨))
          break
    elif int(선택3) == 2:
       self.메뉴화면

    
게임 = MyGame("yb", "사냥꾼")
````
