### **(2) 랜덤값 생성**



####**1. random.random**
* 0과 1사이의 난수를 무작위로 생성합니다.

```
random.random()
```



````python
rnd_float=random.random()
print(rnd_float,type(rnd_float))
````

#### **2. random.randint**
* 특정 범위 랜덤한 정수를 무작위로 생성합니다. (a이상 b 이하)

```
random.randint(a, b)
```



````python
rnd_int=random.randint(1,10)
print(rnd_int,type(rnd_int))
````

### **(3) 랜덤값 추출**

####**1. random.choice**
* 주어진 sequence 에서 랜덤으로 한개의 요소를 뽑습니다.

```
a = [1,2,3]
b = random.choice(a)
print(b)
```



````python
a = [1,2,3]
b = random.choice(a)
print(b)
````

####**2. random.choices**
* 주어진 sequence 에서 랜덤으로 k개의 요소를 뽑습니다. 중복이 가능합니다.

```
a = [1,2,3]
b = random.choices(a, k=2)
print(b)
```



````python
a = [1,2,3]
b = random.choices(a,k=3)  #중복값도 나올수 있기 때문에 set을 활용하던가 choice를 써서 하나씩 뽑던가
print(b)
````

### **(4) 기타**

**1. random.shuffle**
* 특정 시퀀스를 랜덤하게 섞는 함수입니다.

```
a = [1,2,3]
random.shuffle(a)
print(a)
```



````python
a = [1,2,3]
random.shuffle(a)
print(a)
````

####**예제 02**

* 업다운게임
  * 1~100까지 값이 랜덤하게 지정됩니다.
  * 사용자가 답한 답이 랜덤값보다 높으면 "UP"
  * 랜덤값보다 낮으면 "DOWN",
  * 랜덤값을 맞추면 "CORRECT"를 출력합니다.
  * 사용자가 랜덤값을 맞출 때까지 진행됩니다.

````python
def updown():
  answer = random.randint(1,100)
  while True:
    user =int(input("정답입력"))
    if answer < user:
      print("down")
    elif answer > user:
      print("up")
    else:
      print("Correct")
      break

updown()
````

####**예제 03**
* **숫자야구**
  * 0~9까지 값을 지닌 3자리 숫자가 랜덤하게 지정됩니다.
  * 각 자리 숫자는 중복되지 않아야 합니다.

  * 사용자가 지정된 랜덤값을 맞출 때까지 진행됩니다.




* 아래 표는 랜덤값이 123일 때 사용자의 입력과 그에 따른 출력을 나타낸 표입니다. 

입력|출력
---|---
123| 3볼 3홈런
132| 3볼 1홈런
231| 3볼 0홈런 
125| 2볼 2홈런



````python
def num_baseball():
  ran_num=[]
  user_choice=[]
  homerun=0
  ball=0
  while len(ran_num)<3:
    ran_num.append(random.randint(0,9))
    ran_num=set(ran_num)
    ran_num=list(ran_num)
  print(ran_num)
  while homerun<3:
    user = list(input("0-9 입력"))
    for i in range(len(user)):
      user[i] = int(user[i])
    if len(user)>3:
      print("에러 숫자 벗어남")
    elif user in user_choice:
      print("이미 입력한 숫자")
    else:
      homerun = 0
      for i in range(3):
        if user[i] == ran_num[i]:
          homerun += 1
        elif user[i] in ran_num:
          ball +=1
      print("{0}볼 {1}홈런".format(ball,homerun))
     
num_baseball()
````
