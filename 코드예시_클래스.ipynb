
###**(4) 클래스 메소드**

class 안에서 def를 추가 할 수 있습니다. (이를 메소드라 합니다.) Student 클래스 안에 자기 소개하는 메소드를 추가해보겠습니다. 

```
class Student:
    def __init__(self, name, subject)
      self.name = name
      self.subject = subject
      self.level = '3개월차'
    
    def introduce(self):
      print("안녕하세요" self.subject, "수업을 듣는", self.level, self.name , "입니다."

me = Student("이름", "영상보안")
me.introduce()
```

````python
class Student():
    def __init__(self, name, subject):
      self.name = name
      self.subject = subject
      self.level = '3개월차'
    
    def 자기소개(self, intro):
      print("%s 저는 %s이고 %s 과정을 듣고 있습니다."%(intro, self.name, self.subject))
````

````python
import time

class Student():
    def __init__(self, name, subject):
      self.name = name
      self.subject = subject
      self.level = {"Python":0,"데이터분석":0,"딥러닝":0,"C++":0,"리눅스":0,}
    
    def 자기소개(self, intro):
      print("%s 저는 %s이고 %s 과정을 듣고 있습니다."%(intro, self.name, self.subject))

    def 레벨확인(self):
      print("""
          * python skill     : %i
          * 데이터분석 skill : %i   
      """%(self.level["Python"], self.level["데이터분석"]))

    def python_skill_up(self):
      # 수강듣기
      # 수강을 들으시겠습니까? y/n
      대답1 = input("수강을 들으시겠습니까?: ")
      if 대답1 == "y":
        print("수업 듣는중..")
        for i in range(3):
          time.sleep(1)
          print("...")
        대답2 = input("다 들었습니다. 문제를 풀겠습니까?: ")
        if 대답2=="y":
          대답3 = int(input("점수: "))
          if 대답3>=60:
            #skillup
            self.level["Python"]+=1
          

````

````python
me = Student("yb", "스마트웰니스")
````

````python
me.레벨확인()
````

````python
class MyGame():
  def __init__(self, 닉네임, 직업):
    self.닉네임 = 닉네임
    self.직업 = 직업  
    self.체력 = 50 
    self.최대체력 =100

    self.현재장소 = "집"
    self.골드 = 1000
    self.장소리스트 = ["집", "사냥터"]

    self.메뉴화면() # 함수

  def 메뉴화면(self):
    while True:
      print("현재장소:", self.현재장소, "골드:", self.골드, "체력: ", self.체력, "/", self.최대체력)
      print("프로필, 행동, 이동, 게임종료")
      선택 = input(">> 선택하세요: ")
      if 선택=='프로필':
        self.선택_프로필()
      elif 선택=='행동':
        self.선택_행동()
      elif 선택=='이동':
        self.선택_이동()     
      elif 선택=='게임종료':
        break
      else:
        print("잘못 입력하셨습니다.")

  def 선택_프로필(self):
    print("**********************")
    print("* 이름: ", self.닉네임)
    print("* 직업: ", self.직업)
    print("**********************")

  def 선택_이동(self):
    print(self.장소리스트)
    이동장소 = input()
    print(">> ", self.현재장소, "에서 ", 이동장소,"로 이동합니다.")
    self.현재장소=이동장소

  def 선택_행동(self):
    if self.현재장소=='집':
      self.장소_집()
    else: 
      print(">> 잘못 입력하였습니다.")
  
  def 장소_집(self): 
    선택 = input(">> 행동을 선택하세요: 1.휴식 2.취침 3.책읽기")
    if int(선택)==1:
      self.체력=min(self.체력+10, self.최대체력)
      print(">> 체력이 10 회복 하였습니다.")
    elif int(선택)==2:
      self.체력=min(self.체력+100, self.최대체력)
      print(">> 체력이 100 회복 하였습니다.")
    else:
      print(">> 잘못 입력하였습니다.")

    
게임 = MyGame("yb", "사냥꾼")
````
